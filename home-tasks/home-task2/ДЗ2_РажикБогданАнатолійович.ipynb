## –†–µ–∑—É–ª—å—Ç–∞—Ç–∏:

- –í–∞—Ä—Ç—ñ—Å—Ç—å –≥–æ–¥–∏–Ω–∏ —Ä–æ–∑—Ä–æ–±–Ω–∏–∫–∞ (y): 25 –¥–æ–ª–∞—Ä—ñ–≤

```python
import numpy as np
A = np.array([
    [-1, 1, 2],
    [0, -1, -3],
    [4, -3, 2]
    ])
B = np.array(
    [1, -4, 7]
)
```


- –ù–∞–ø–∏—à–∏ —Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è –≤–∏—Ä—ñ—à–µ–Ω–Ω—è —Å–∏—Å—Ç–µ–º–∏ –º–∞—Ç—Ä–∏—á–Ω–∏–º –º–µ—Ç–æ–¥–æ–º
- –ù–∞–ø–∏—à–∏ —Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è –≤–∏—Ä—ñ—à–µ–Ω–Ω—è —Å–∏—Å—Ç–µ–º–∏ –º–µ—Ç–æ–¥–æ–º –ö—Ä–∞–º–µ—Ä–∞

```python
# –ù–∞–ø–∏—à–∏ —Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è –≤–∏—Ä—ñ—à–µ–Ω–Ω—è —Å–∏—Å—Ç–µ–º–∏ –º–∞—Ç—Ä–∏—á–Ω–∏–º –º–µ—Ç–æ–¥–æ–º
def solve_inv_matrix(a, b, verbose=False):
    try:
        a_inv = np.linalg.inv(a)
        x = np.dot(a_inv, b)
        return x
    except np.linalg.LinAlgError as e:
        return f"–ü–æ–º–∏–ª–∫–∞: {e}"

print(f"–í–µ–∫—Ç–æ—Ä —Ä—ñ—à–µ–Ω–Ω—è: \r\n {solve_inv_matrix(A, B)}")
```

```python
# –ù–∞–ø–∏—à–∏ —Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è –≤–∏—Ä—ñ—à–µ–Ω–Ω—è —Å–∏—Å—Ç–µ–º–∏ –º–µ—Ç–æ–¥–æ–º –ö—Ä–∞–º–µ—Ä–∞
def solve_cramer(a, b, verbose=False):
    try:
        det_a = np.linalg.det(a)
        if np.isclose(det_a, 0):
            return "–°–∏—Å—Ç–µ–º–∞ –Ω–µ –º–∞—î —î–¥–∏–Ω–æ–≥–æ —Ä–æ–∑–≤'—è–∑–∫—É (–≤–∏–∑–Ω–∞—á–Ω–∏–∫ = 0)."

        n = len(b)
        x = np.zeros(n)
        for i in range(n):
            a_temp = a.copy()
            a_temp[:, i] = b
            det_a_temp = np.linalg.det(a_temp)
            x[i] = det_a_temp / det_a
        return x
    except np.linalg.LinAlgError as e:
        return f"–ü–æ–º–∏–ª–∫–∞: {e}"

print(f"–í–µ–∫—Ç–æ—Ä —Ä—ñ—à–µ–Ω–Ω—è: \r\n {solve_cramer(A, B)}")
```
