Виконай диференціювання функції та знайди значення похідних у точках



```python
from sympy import symbols, diff, sqrt

x = symbols('x')

f1 = x**3 / 3 + x**2 / 2 - 2 * x
f2 = sqrt(x**2 + 1)
f3 = 1 / sqrt(x**2 + 1)

f1_prime = diff(f1, x)
f2_prime = diff(f2, x)
f3_prime = diff(f3, x)

points = [1, -1/2]

print("Результати обчислень:")
for func_name, derivative in [
    ("f1'(x)", f1_prime),
    ("f2'(x)", f2_prime),
    ("f3'(x)", f3_prime)
]:
    print(f"\n{func_name}: {derivative}")
    for point in points:
        value = derivative.subs(x, point).evalf()
        print(f"  {func_name} в точці x = {point}: {value}")
```

Завдання 2

Аналітично виконай диференціювання функції та знайди точку максимуму. Програмно візуалізуй функцію, її похідну та точку максимуму.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, solve

x = symbols('x')

y = -3 * x**2 + 30 * x

y_prime = diff(y, x)

critical_points = solve(y_prime, x)
critical_values = {point: y.subs(x, point) for point in critical_points}

def y_func(x):
    return -3 * x**2 + 30 * x

def y_prime_func(x):
    return -6 * x + 30

x_vals = np.linspace(0, 10, 500)
y_vals = y_func(x_vals)
y_prime_vals = y_prime_func(x_vals)

x_max = float(critical_points[0])
y_max = float(critical_values[int(x_max)])

plt.figure(figsize=(10, 6))

plt.plot(x_vals, y_vals, label="$y(x) = -3x^2 + 30x$", color="blue")
plt.plot(x_vals, y_prime_vals, label="$y'(x)$", color="orange", linestyle="--")
plt.scatter([x_max], [y_max], color="red", label=f"Максимум ({x_max}, {y_max})", zorder=5)

plt.axhline(0, color="black", linewidth=0.7, linestyle="--")
plt.axvline(x_max, color="gray", linewidth=0.7, linestyle="--")
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

print("Аналітичне розв'язання:")
print(f"Функція: {y}")
print(f"Похідна: {y_prime}")
print(f"Критична точка (максимум): x = {x_max}, y = {y_max}")
```
