## Завдання 3

Підкидають три гральні кубики. Що ймовірніше: отримати в сумі очок, що випали, 11 або 12?  
Розв'яжи програмно, згенерувавши всі можливі комбінації.

---
Розв'язання:

````python
import itertools

outcomes = list(itertools.product(range(1, 7), repeat=3))
sum_11 = sum(1 for outcome in outcomes if sum(outcome) == 11)
sum_12 = sum(1 for outcome in outcomes if sum(outcome) == 12)
print(f"Ймовірність суми 11: {sum_11 / len(outcomes):.4f}")
print(f"Ймовірність суми 12: {sum_12 / len(outcomes):.4f}")
````

## Завдання 7. Випадкове блукання
Ціна акцій компанії в кожен момент часу може з рівною ймовірністю збільшитись на 2 або зменшитись на 1. В початковий момент часу $t=0$ ціна рівна 0. Визнач середню ціну акції через 3 одиниці часу ($t=3$).  
а) Виріши аналітично.

б) Напиши симуляцію данного процесу. Порахуй середнє значення та намалюй гістограму ціни для 10, 100, 1000 та 10000 симуляцій.

---
Розв'язання:
photo_2024-12-04_10-53-07.jpg

````python
import numpy as np

def stock_price_at_time(t):
    price = 0
    for _ in range(t):
        step = np.random.choice([2, -1])
        price += step
    return price

def simulate_n_times(n: int, t: int) -> list:
    np.random.seed(42)
    results = [stock_price_at_time(t) for _ in range(n)]
    return results
````

````python
import matplotlib.pyplot as plt
for n in [10, 100, 1000, 10000]:
    prices = simulate_n_times(n, 3)
    mean_price = np.mean(prices)

    print(f"Середнє значення ціни для {n} симуляцій: {mean_price:.2f}")

    plt.figure(figsize=(8, 4))
    plt.hist(prices, bins=20, color='blue', alpha=0.7, edgecolor='black')
    plt.title(f"Гістограма цін (n={n})")
    plt.xlabel("Ціна")
    plt.ylabel("Частота")
    plt.grid(alpha=0.3)
    plt.show()
````
