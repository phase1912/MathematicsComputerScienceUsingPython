#Завдання 1:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, normaltest

def stock_price_at_time(t):
    price = 0
    for _ in range(t):
        step = gamma.rvs(a=0.3, scale=1.1)
        price += step
    return price

def simulate_n_times(n, t):
    np.random.seed(42)
    results = [stock_price_at_time(t) for _ in range(n)]
    return results

# Praams
n = 100  # Count
time_steps = range(1, 61, 2)

# Gistogram
single_step_data = simulate_n_times(n, 1)
plt.figure(figsize=(8, 4))
plt.hist(single_step_data, bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.title("Гістограма розподілу зміни ціни за один крок")
plt.xlabel("Зміна ціни")
plt.ylabel("Частота")
plt.grid(alpha=0.3)
plt.show()

for t in time_steps:
    prices = simulate_n_times(n, t)
    stat, p_value = normaltest(prices)

    print(f"t={t}: p-value={p_value:.4f} ({'нормальний' if p_value > 0.05 else 'ненормальний'} розподіл)")

    plt.figure(figsize=(8, 4))
    plt.hist(prices, bins=20, color='green', alpha=0.7, edgecolor='black')
    plt.title(f"Гістограма ціни акцій (t={t})")
    plt.xlabel("Ціна")
    plt.ylabel("Частота")
    plt.grid(alpha=0.3)
    plt.show()

print("\nВисновки:")
print("1. Зі збільшенням часу t розподіл стає більш симетричним.")
print("2. При достатньо великому t розподіл наближається до нормального (згідно з тестом на нормальність).")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, pearsonr

url = 'https://docs.google.com/spreadsheets/d/18WCpPS96Tb3cB0FCsIA92PEhcmBkp08sjYhS9DsQfJE/edit#gid=954244094'
url = url[:url.find('/edit')] + '/export?format=csv'
df = pd.read_csv(url)

def analyze_column(col_name):
    col = df[col_name]
    mean = col.mean()
    variance = col.var()
    std_dev = col.std()

    stat, p_value = shapiro(col)
    normality = p_value > 0.05

    corr, _ = pearsonr(col, df["Product_Sold"])

    plt.figure(figsize=(8, 4))
    sns.histplot(col, kde=True, bins=10, color='skyblue', edgecolor='black')
    plt.title(f"Гістограма {col_name}")
    plt.xlabel(col_name)
    plt.ylabel("Частота")
    plt.show()

    return {
        "mean": mean,
        "variance": variance,
        "std_dev": std_dev,
        "normality_p_value": p_value,
        "is_normal": normality,
        "correlation_with_Product_Sold": corr,
    }

results = {}
for col in df.columns:
    if col != "Product_Sold":
        print(f"=== Аналіз для {col} ===")
        results[col] = analyze_column(col)
        print(results[col])

print("\nРезультати аналізу:")
for key, value in results.items():
    print(f"{key}: {value}")
```
